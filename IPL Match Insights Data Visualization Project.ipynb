```python
'''
1.Right now, the date column is stored as a plain text (object).
2.Python canâ€™t understand, sort, or filter it like a date.
3.When you convert it to datetime, Python understands that it's a real date and time.
'''
df['date'] = pd.to_datetime(df['date'])
```

```python
'''
1.These columns contain repeated string values (e.g., "Chennai Super Kings", "win", "field").
2.Storing the same text 1000+ times wastes memory.
3.category stores the unique values once and replaces the rest with codes internally.
'''
cat_columns = [
    'season', 'city', 'match_type', 'team1', 'team2',
    'toss_winner', 'toss_decision', 'winner',
    'result', 'super_over', 'method', 'umpire1', 'umpire2'
]

for col in cat_columns:
    df[col] = df[col].astype('category')
```

```python
'''
| Column          | Description                                  | Recommended? | Why                                                             |
| --------------- | -------------------------------------------- | ------------ | --------------------------------------------------------------- |
| `target_runs`   | Target runs for the second team              | âœ… Yes        | Runs are always whole numbers                                   |
| `result_margin` | Win margin (runs or wickets)                 | âœ… Yes        | Always whole numbers                                            |
| `target_overs`  | Number of overs in the second innings target | âŒ Maybe not  | Could be **fractional** (e.g., `17.5`) in rain-affected matches |

'''
df['target_runs'] = df['target_runs'].fillna(0).astype(int)
df['result_margin'] = df['result_margin'].fillna(0).astype(int)
```

```python
df.isnull().sum()
```

```python
# Step 1: Fill 'target_overs' (numeric column)
df['target_overs'].fillna(0, inplace=True)

# Step 2: Drop the 'method' column since it's all missing
df.drop(columns='method', inplace=True)

# Step 3: Handle missing categorical values

# Define column-value pairs for filling
fill_map = {
    'player_of_match': 'Unknown',
    'winner': 'No Result',
    'city': 'Unknown'
}

for col, missing_value in fill_map.items():
    # Convert column to 'category' if not already
    if df[col].dtype != 'category':
        df[col] = df[col].astype('category')

    # Add missing value label if not present in categories
    if missing_value not in df[col].cat.categories:
        df[col] = df[col].cat.add_categories([missing_value])

    # Fill missing values
    df[col] = df[col].fillna(missing_value)
```

```python
#1. Top 10 Cities that Hosted the Most Matches (Bar Plot)
plt.figure(figsize=(10, 5))
city_order = df['city'].value_counts().head(10)
sns.barplot(x=city_order.values, y=city_order.index, palette='viridis')
plt.title('Top 10 Cities That Hosted the Most Matches')
plt.xlabel('Number of Matches')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
#2. Top 10 Venues by Match Count (Horizontal Bar Plot)
plt.figure(figsize=(10, 6))
venue_order = df['venue'].value_counts().head(10)
sns.barplot(x=venue_order.values, y=venue_order.index, palette='magma')
plt.title('Top 10 Venues by Match Count')
plt.xlabel('Number of Matches')
plt.ylabel('Venue')
plt.tight_layout()
plt.show()
```

```python
# 3. Top 10 Teams with Most Wins (Bar Plot)
plt.figure(figsize=(10, 6))
top_winners = df['winner'].value_counts().head(10)
sns.barplot(x=top_winners.values, y=top_winners.index, palette='coolwarm')
plt.title('Top 10 Teams with Most Wins')
plt.xlabel('Number of Wins')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

```python
#4. Toss Decision Distribution (Pie Chart)
plt.figure(figsize=(6, 6))
df['toss_decision'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen'])
plt.title('Toss Decision Distribution')
plt.ylabel('')
plt.show()
```

```python
# 5. Match Result Types (Normal/ Super Over / No Result / Tie) (Count Plot)
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='result', palette='Set1')
plt.title('Distribution of Match Results')
plt.xlabel('Result Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#6. Top 10 Most Frequently Appearing Umpires (Bar Plot)
plt.figure(figsize=(10, 6))
umpires = pd.concat([df['umpire1'], df['umpire2']])
top_umpires = umpires.value_counts().head(10)
sns.barplot(x=top_umpires.values, y=top_umpires.index, palette='pastel')
plt.title('Top 10 Umpires by Appearances')
plt.xlabel('Number of Matches')
plt.ylabel('Umpire')
plt.tight_layout()
plt.show()
```

```python
#7.12. Average Target Runs Per Season (Line Plot)
avg_runs_per_season = df.groupby('season')['target_runs'].mean()

plt.figure(figsize=(10, 5))
sns.lineplot(x=avg_runs_per_season.index, y=avg_runs_per_season.values, marker='o', color='darkgreen')
plt.title('Average Target Runs Per Season')
plt.xlabel('Season')
plt.ylabel('Average Target Runs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
